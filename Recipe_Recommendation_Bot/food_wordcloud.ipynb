```python
# Program to generate Word Cloud

# import all necessary modules
from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
```

```python
# Read CSV File
import os
df = pd.read_csv("foodie_words.csv", on_bad_lines="skip")
text = " ".join(df["asparagus"])

text
```

```python
# read the mask/color image taken from
#coloring = np.array(Image.open('.jpg'))
stopwords = set(STOPWORDS)
custom_stop_words = [""]
[stopwords.add(n) for n in custom_stop_words]
```

```python
wc = WordCloud(background_color="white" , width=500, max_words=3500, 
              stopwords=stopwords, max_font_size=100, random_state=42)

#generate word cloud
wc.generate(text)
fontdict={"fontsize": 48, "fontweight":"bold"}
plt.title("Ingredients Word Cloud", fontdict=fontdict)
plt.imshow(wc) 
```
